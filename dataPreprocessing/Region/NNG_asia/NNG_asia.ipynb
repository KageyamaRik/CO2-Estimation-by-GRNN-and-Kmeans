```python
import geopandas as gpd
import numpy as np
from shapely.geometry import box, Polygon
from shapely.ops import orient
from pyproj import Geod
from maplot import *

world_shp = gpd.read_file("H:\\Region\\世界国家\\世界国家.shp")
world_shp
```

```python
countries = {"CHINA", "MONGOLIA", "KOREA,DEMOCRATIC PEOPLE'S REPUBLIC OF", "KOREA, REPUBLIC OF", "JAPAN",  # East Asia
             "MYANMAR", "THAILAND", "LAOS", "CAMBODIA", "VIET NAM", "PHILIPPINES", "MALAYSIA", "INDONESIA",  # Southeast Asia
             "KAZAKHSTAN", "KYRGYZSTAN", "TAJIKISTAN", "UZBEKISTAN", "TURKMENISTAN",  # Central Asia
             "RUSSIAN FEDERATION"  # North Asia (Siberia)
             }
region_shp = world_shp[world_shp["NAME"].isin(countries)]
region_shp.plot()
```

## 草

```python
grids = np.full((180, 360), False)

for lat in range(180):
    for lon in range(360):
        bbox_polygon = box(lon - 180., lat - 90., lon - 179., lat - 89.)
        bbox_gdf = gpd.GeoDataFrame(geometry=[bbox_polygon], crs=region_shp.crs)
        clipped_gdf = gpd.clip(region_shp, bbox_gdf)
        clipped_gdf['area'] = clipped_gdf.geometry.area

        if clipped_gdf['area'].sum() > 0.4:
            grids[lat, lon] = True
```

```python
longitude = np.linspace(-179.5, 179.5, 360)
latitude = np.linspace(-89.5,89.5,180)
lon, lat = np.meshgrid(longitude, latitude)

plt.rcParams['figure.figsize'] = (12.0, 8.0)
Map = Basemap(llcrnrlon=45., llcrnrlat=-12., urcrnrlon=160., urcrnrlat=80.)
Map.drawcoastlines()
parallels = np.arange(-75.,76,25.)
# labels = [left,right,top,bottom]
Map.drawparallels(parallels,labels=[True,False,False,False])
meridians = np.arange(0.,361.,60.)
Map.drawmeridians(meridians,labels=[False,False,False,True])

x, y = Map(lon, lat)

mapPlot = Map.pcolor(x, y, grids, cmap='tab10')
mapPlot.cmap.set_under('w')  # 设置底色
plt.show()
```
