```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
from matplotlib.ticker import MaxNLocator
from mpl_toolkits.basemap import Basemap

inRoot = f"H:\\XCO2\MappingXCO2\\new_lat_interpo\\MappingXCO2_5%_bkg_ano_year\\"
china_shp_path = ("H:\\Region\\区划\\省", "省")  # 中国陆地轮廓
nine_shp_path = ("H:\\Region\\全国shp\\最新2021年全国行政区划\\九段线", "九段线")  # 九段线
```

```python
# 绘制测试
dataTest = np.load(inRoot + "MappingXCO2_2014_lat_5%_ano_year.npy")
dataTest
```

```python
longitude = np.linspace(-179.5, 179.5, 360)
latitude = np.linspace(-89.5, 89.5, 180)
lon, lat = np.meshgrid(longitude, latitude)

plt.rcParams['figure.figsize'] = (12.0, 8.0)
Map = Basemap(llcrnrlon=70., llcrnrlat=16., urcrnrlon=140., urcrnrlat=55.)
parallels = np.arange(-75., 76, 25.)

# labels = [left,right,top,bottom]
Map.drawparallels(parallels, labels=[True, False, False, False])
meridians = np.arange(0., 361., 60.)
Map.drawmeridians(meridians, labels=[False, False, False, True])
Map.readshapefile(*china_shp_path)
Map.readshapefile(*nine_shp_path)

x, y = Map(lon, lat)

# 需要翻转！
mapPlot = Map.pcolor(x, y, np.flip(dataTest, axis=0), cmap="viridis")
mapPlot.cmap.set_under('w')  # 设置底色
plt.show()
```

```python
for year in range(2014, 2021):
    data = np.load(inRoot + f"MappingXCO2_{year}_lat_5%_ano_year.npy")
    np.save(f"H:\\XCO2\\MappingXCO2\\MappingXCO2_latitude_interp_5percent_year\\MappingXCO2_{year}", np.flip(data, axis=0))
```
